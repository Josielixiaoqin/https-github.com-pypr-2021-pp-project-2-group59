```python
import pandas as pd
dft = pd.read_csv('dft_rawcount_region_id_3.csv')
cdata=dft[(dft['road_type']=='Major') & (dft['local_authority_name']=='City of Edinburgh')]##做一下关于'hour'和各种交通方式''的分布
data=pd.DataFrame(cdata,columns=["latitude","longitude","pedal_cycles"])
data.head()
```

```python
import os
import folium
import numpy as np
from folium.plugins import HeatMap

m = folium.Map([ 55.96, -3.307], tiles='stamentoner', zoom_start=11)

HeatMap(data).add_to(m)

m.save(os.path.join(r'C:\Users\hp\Desktop', 'Heatmap.html'))

m
```
